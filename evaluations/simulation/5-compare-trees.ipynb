# 1. Compare trees

## 1.1. Compare reads SNP tree

```r
library(ape)
library(phytools)
library(phangorn)
library(repr)

read_and_root_tree <- function(file) {
    tree <- read.tree(file=file)
    tree <- root(tree, "reference", resolve.root=TRUE)
    
    return(tree)
}

compare_trees <- function(tree1, tree2, case) {
    print(paste(case, ": comparePhylo", sep=""))
    print(comparePhylo(tree1, tree2, plot=FALSE))
    print(paste(case, ": treeDist", sep=""))
    print(treedist(tree1, tree2, check.labels = TRUE))
    print(paste(case, ": cophylo", sep=""))
    
    options(repr.plot.width=10, repr.plot.height=16)
    return(plot(cophylo(tree1, tree2, rotate=TRUE)))
}

t_orig <- read_and_root_tree(file="input/salmonella.tre")

t_snp_reads <- read_and_root_tree(file="index-reads/reads.snp-tree.tre")
compare_trees(t_orig, t_snp_reads, "Compare original to reads (SNP)")
```

## 1.2. Compare reads SNP and MNP tree

```r
t_snp_mnp_reads <- read_and_root_tree(file="index-reads/reads.snp-mnp-tree.tre")
compare_trees(t_orig, t_snp_mnp_reads, "Compare original to reads (SNP and MNP)")
```

# 2. Compare assemblies

## 2.1. Compare assemblies SNP tree

```r
t_snp_assemblies <- read_and_root_tree(file="index-assemblies/assemblies.snp-tree.tre")
compare_trees(t_orig, t_snp_assemblies, "Compare original to assemblies (SNP)")
```

## 2.2. Compare assemblies SNP and MNP tree

```r
t_snp_mnp_assemblies <- read_and_root_tree(file="index-assemblies/assemblies.snp-mnp-tree.tre")
compare_trees(t_orig, t_snp_mnp_assemblies, "Compare original to assemblies (SNP and MNP)")
```
