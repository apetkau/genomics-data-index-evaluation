# 1.1. Compare reads

```r
library(ape)
library(phytools)
library(phangorn)
library(repr)

t_orig <- read.tree(file="input/salmonella.tre")
t_new <- read.tree(file="index-reads/reads-tree.tre")

t_orig <- root(t_orig, "reference", resolve.root=TRUE)
t_new <- root(t_new, "reference", resolve.root=TRUE)

comparePhylo(t_orig, t_new, plot=TRUE)
```

```r
treedist(t_orig, t_new, check.labels = TRUE)
```

```r
options(repr.plot.width=15, repr.plot.height=16)

plot(cophylo(t_orig, t_new, rotate=TRUE))
```
