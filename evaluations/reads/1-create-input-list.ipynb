# 1. Load metadata

```python
import pandas as pd

columns_to_keep = ['dataSetName', 'strain', 'outbreak', 'suggestedReference']

campy_df = pd.read_csv('data/datasets/datasets/Campylobacter_jejuni_0810PADBR-1.tsv', sep='\t', skiprows=8)
campy_df = campy_df.loc[:, columns_to_keep]
campy_df = campy_df[~campy_df['strain'].isna()]
campy_df
```

```python
ecoli_df = pd.read_csv('data/datasets/datasets/Escherichia_coli_1405WAEXK-1.tsv', sep='\t', skiprows=7)
ecoli_df = ecoli_df.loc[:, columns_to_keep]
ecoli_df = ecoli_df[~ecoli_df['strain'].isna()]
ecoli_df
```

```python
listeria_df = pd.read_csv('data/datasets/datasets/Listeria_monocytogenes_1408MLGX6-3WGS.tsv', sep='\t', skiprows=8)
listeria_df = listeria_df.loc[:, columns_to_keep]
listeria_df = listeria_df[~listeria_df['strain'].isna()]
listeria_df
```

```python
salmonella_df = pd.read_csv('data/datasets/datasets/Salmonella_enterica_1203NYJAP-1.tsv', sep='\t', skiprows=7)
salmonella_df = salmonella_df.loc[:, columns_to_keep]
salmonella_df = salmonella_df[~salmonella_df['strain'].isna()]
salmonella_df
```

```python
metadata_df = pd.concat([campy_df, ecoli_df, listeria_df, salmonella_df])
metadata_df['strain'] = metadata_df['strain'].str.strip()
metadata_df
```

```python
metadata_df.to_csv('metadata.tsv', sep='\t', index=False)
```

# 2. Create reads input files

```python
from pathlib import Path

input_df = metadata_df.copy()
input_df['Sample'] = input_df['strain']
input_df['Assemblies'] = pd.NA
input_df['Reads1'] = input_df['Sample'].apply(lambda x: str(Path('data') / 'fastq' / (x + '_1.fastq.gz')))
input_df['Reads2'] = input_df['Sample'].apply(lambda x: str(Path('data') / 'fastq' / (x + '_2.fastq.gz')))
input_df = input_df[['Sample', 'Assemblies', 'Reads1', 'Reads2']]
input_df
```

```python
input_df.to_csv('input.tsv', sep='\t', index=False)
```
